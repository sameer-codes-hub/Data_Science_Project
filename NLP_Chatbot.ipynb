```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
ps=PorterStemmer()
lemma=nltk.WordNetLemmatizer()
stop=set(stopwords.words('english'))
```

```python
nltk.download('punkt_tab')
dt['answer']=dt['answer'].astype(str)
dt['toke']=dt['answer'].apply(word_tokenize)
```

```python
dt.head()
```

```python
dt['stop_word']=dt['toke'].apply(lambda x:[word.lower()for word in x if word.lower() not in stop and word.isalpha()])
```

```python
dt['stem']=dt['stop_word'].apply(lambda x:[ps.stem(word) for word in x])
```

```python
dt.head()
```

```python
dt['lemma']=dt['stop_word'].apply(lambda x:[lemma.lemmatize(word) for word in x])
```

```python
dt['processed_text']=dt['lemma'].apply(lambda x:" ".join(x))
```

```python
vect=CountVectorizer()
vec=vect.fit_transform(dt['processed_text'])
```

```python
corp=dt['processed_text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tdf=TfidfVectorizer(max_features=5000)
x=tdf.fit_transform(corp)
```

```python
x.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def medical_chatbot(user_question):
  user_question=user_question.lower()
  user_vec=tdf.transform([user_question])
  similarity=cosine_similarity(user_vec,x)
  score=similarity.max()
  if score < 0.2:
    return "Sorry I don't have information about this question."
  index=similarity.argmax()
  return dt.iloc[index] ['answer']
```

```python
medical_chatbot("What are the treatments for Glaucoma ?	")
```

```python
medical_chatbot("what can i use for headache")
```
