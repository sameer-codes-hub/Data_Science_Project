```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
ps=PorterStemmer()
lemma=nltk.WordNetLemmatizer()
stopwords=set(stopwords.words('english'))
```

```python
nltk.download('punkt_tab')
df['answer'] = df['answer'].astype(str)
df['toke']=df['answer'].apply(word_tokenize)
```

```python
df.head()
```

```python
df['stop_word']=df['toke'].apply(lambda x:[word.lower()for word in x if word.lower() not in stopwords and word.isalpha()])
```

```python
df['stem']=df['stop_word'].apply(lambda x:[ps.stem(word) for word in x])
```

```python
df.head()
```

```python
df['lemma']=df['stop_word'].apply(lambda x:[lemma.lemmatize(word) for word in x])
```

```python
df['processed_text']=df['lemma'].apply(lambda x:" ".join(x))
```

```python
vec=CountVectorizer()
v=vec.fit_transform(df['processed_text'])
```

```python
df.head()
```

```python
word=vec.get_feature_names_out()
```

```python
back_of_word=pd.DataFrame(v.toarray(),columns=word)
```
